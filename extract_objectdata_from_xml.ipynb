```python
import os
import pandas as pd
from glob import glob
from functools import reduce
from xml.etree import ElementTree as et

print("All imports successful âœ…")
```

```python
# Load all xml files and store in a list
xml_list = glob('./data/data_images/*.xml')
```

```python
xml_list
```

```python
# Read xml file
# From each file extract - filename, size(width, height), object(name, xmin, xmax, ymin, ymax)
def extract_text(filename):
    tree = et.parse(filename)
    root = tree.getroot()
    # print(root)
    # output - <Element 'annotation' at 0x10dd4e770>

    # Extract file name
    image_name = root.find('filename').text

    # Extract width and height of the image
    width = root.find('size').find('width').text
    height = root.find('size').find('height').text

    # Extract object(name, xmin, xmax, ymin, ymax)
    objs = root.findall('object')
    parser = []
    for obj in objs:
        name = obj.find('name').text
        bndbox = obj.find('bndbox') # Finding boundingbox
        xmin = bndbox.find('xmin').text
        xmax = bndbox.find('xmax').text
        ymin = bndbox.find('ymin').text
        ymax = bndbox.find('ymax').text
        parser.append([image_name, width, height, name, xmin, xmax, ymin, ymax])
        
    return parser
```

```python
parser_all = list(map(extract_text, xml_list))
```

```python
parser_all # now this is in a multidimensional format to inorder to flatten it we use reduce function``
```

```python
data = reduce(lambda x, y : x+y, parser_all)
```

```python
data # now data is in a 2d form where we have multiple rows and columns
```

```python
df = pd.DataFrame(data, columns = ['filename', 'width', 'height', 'obj_name', 'xmin', 'xmax', 'ymin', 'ymax'])
```
